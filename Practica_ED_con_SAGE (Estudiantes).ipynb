# Ecuaciones Diferenciales con condiciones iniciales

## En Sage, las condiciones iniciales se agregan usando el par√°metro ics=[x0, y0] dentro de "desolve".

## por ejemplo, si la ecuaci√≥n es: $\frac{dy}{dx} = y$ 

## La soluci√≥n general es: $ùë¶(x) = Ce^x$

## Si se quiere la condici√≥n inicial: $y(0)=2$ 

## Entonces:

```python
var('x')
y = function('y')(x)

sol = desolve(diff(y,x) - y == 0, y, ics=[0,2])
sol
```
