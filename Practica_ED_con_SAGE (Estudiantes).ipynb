# - **Ejercicio 1:** calcular el area bajo la curva de las funciones: $f(x)=x^2$ y g(x)=2x+1

```python
f(x)=x^2
g(x)=2*x+1

# intersecciones
a,b = [s.rhs() for s in solve(f(x)==g(x), x)]

# curvas
p1 = plot(f,(x,a,b),color='blue',legend_label='$f(x)=x^2$')
p2 = plot(g,(x,a,b),color='red',legend_label='$g(x)=2x+1$')

# sombreado entre curvas (truco Sage)
area_plot = plot(g,(x,a,b), fill=f, color='green', alpha=0.3)

show(p1+p2+area_plot)
```

# Ecuaciones homogéneas

## No todas las ecuaciones diferenciales de primer orden son de variables separables, pero algunas de ellas son reducibles a variables separables,como es el caso de las homogéneas. Ya sabemos que estas ecuaciones son de la forma $y' = f (y/x)$. También aquı́ podemos usar la función "desolve( )"

# 📘 Ecuación diferencial homogénea

## Resolvamos la ecuación diferencial:
$$
y' = \frac{x+y}{x}
$$
## Observa que podemos separarla:
$$
\frac{x+y}{x} = \frac{x}{x} + \frac{y}{x}
$$
## Por lo tanto:
$$
y' = 1 + \frac{y}{x}
$$
## Ahora notamos que el lado derecho depende **únicamente del cociente**:
$$
\frac{y}{x}
$$
## Entonces la ecuación es **homogénea de primer orden**, ya que tiene la forma general:
$$
y' = F\!\left(\frac{y}{x}\right)
$$

# 🔹 Método de solución

## Hacemos el cambio de variable:

$$
z = \frac{y}{x}, \qquad y = zx
$$

## Derivando:

$$
y' = z + x z'
$$

## Sustituyendo en la ecuación original:

$$
z + x z' = 1 + z
$$

## Restando \(z\):

$$
x z' = 1
$$

## Finalmente:

$$
z' = \frac{1}{x}
$$

## Integramos:

$$
z = \ln|x| + C
$$

## Regresando a \(y\):

$$
y = x(\ln|x| + C)
$$
